```python
#Common friends finder in sets 
your_friends = {"rahul", "amit", "sneha", "priya"}
his_friends = {"amit", "priya", "rohan", "neha"}

# Common friends
common = your_friends & his_friends

# All friends
all_friends = your_friends | his_friends

# Only your friends
only_yours = your_friends - his_friends

print("Common friends:", common)
print("All friends:", all_friends)
print("Only your friends:", only_yours)
```

```python
student = {"name": "Zia", "math": 90, "python": 85, "english": 70}

total = student["math"] + student["python"] + student["english"]
percentage = total / 3

# Pass/Fail condition (each subject >= 40)
if student["math"] >= 40 and student["python"] >= 40 and student["english"] >= 40:
    result = "Pass ‚úÖ"
else:
    result = "Fail ‚ùå"

print("Name:", student["name"])
print("Total Marks:", total)
print("Percentage:", percentage)
print("Result:", result)
```

```python
sentence = input("Enter a sentence: ")
words = sentence.split()
freq = {}
for w in words:
    w = w.lower()   # same word ko same count karne ke liye
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1
print("Word Frequency:", freq)
```

```python
lines = []
for i in range(1, 4):
    line = input(f"Enter line {i}: ")
    lines.append(line + "\n")

# Write to file
with open("notes.txt", "w") as file:
    file.writelines(lines)

print("\n‚úÖ Notes saved successfully!\n")

# Read and print file
with open("notes.txt", "r") as file:
    content = file.read()

print("üìÑ Notes from file:")
print(content)
```
