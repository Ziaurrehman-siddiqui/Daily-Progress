```python
#Task 9: Count Vowels in a Word
word=input("Enter a Word")
count=0
for ch in word:
    if ch.lower() in "aeiou":
        count+=1

print("total vowels",count)
```

```python
#Task 10: Student Marks List
marks=[]

for i in range(1,6):
    m=int(input(f"Enter marks {i}:"))
    marks.append(m)

print("Marks List",marks)
print("Maximum Marks",max(marks))
print("Minimum Marks",min(marks))
print("Average Marks",sum(marks)/len(marks))
```

```python
# Remove Duplicates from list
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print("Original list:", numbers)
print("Unique list:", unique_numbers)
```

```python
# take item from user to add in cart and stop if user give done
cart = []
while True:
    item = input("Enter item (type 'done' to stop): ")
    if item.lower() == "done":
        break
    cart.append(item)
print("\nðŸ›’ Your Shopping Cart:", cart)
print("Total items:", len(cart))
```
