```python
print(" Todo App (type 'done' to stop)")

with open("todo.txt", "a") as file:
    while True:
        task = input("Enter a task: ")

        if task.lower() == "done":
            break

        file.write(task + "\n")

print("\n✅ Tasks saved!\n")

# Show all tasks from file
print(" Your Todo List:")
with open("todo.txt", "r") as file:
    print(file.read())
```
